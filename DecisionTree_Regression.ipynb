```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age','bmi','children','sex_male','smoker_yes']]
independent
```

```python
dependent=dataset['charges']
dependent
```

```python
# here sklearn library used to split train and test model
from sklearn.model_selection import train_test_split
#train_test_split function
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion="mae",splitter="random",max_features="sqrt")
regressor=regressor.fit(X_train,Y_train)
```

```python
#this library for graph purpose
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
#to find r2_score,we take actual output and predictor.
r_score=r2_score(Y_test,y_pred)
```

```python
r_score
```

```python
#pickle is used to save model creation
import pickle
#creating filename using extension.sav
filename="finalized_model_decisiontree.sav"
```

```python
pickle.dump(regressor,open(filename,"wb"))
```

```python
#load the model and rb is used for just read 
loaded_model=pickle.load(open("finalized_model_decisiontree.sav","rb"))
#we check it and for prediction we can do it
result=loaded_model.predict([[1234,345,4565,567,894]])
```
